```python
import open3d as o3d
import numpy as np

def load_multiple_pcd_frames(file_path):
    with open(file_path, 'r') as f:
        content = f.read().strip()

    # Split frames based on 'VERSION' keyword
    frames = content.split("VERSION 0.7")[1:]  # Ignore first split since it's empty

    point_clouds = []
    for frame in frames:
        lines = frame.strip().split("\n")
        
        # Extract only numerical data
        data_lines = []
        is_data_section = False
        for line in lines:
            if line.startswith("DATA"):
                is_data_section = True
                continue
            if is_data_section and line.strip():
                data_lines.append(line)

        # Convert data to NumPy array if not empty
        if len(data_lines) > 0:
            try:
                points = np.loadtxt(data_lines, dtype=np.float32)
                if points.ndim == 1:  # Fix case where only one point exists
                    points = points.reshape(1, -1)
                if points.shape[1] >= 3:  # Ensure at least x, y, z
                    point_cloud = o3d.geometry.PointCloud()
                    point_cloud.points = o3d.utility.Vector3dVector(points[:, :3])  # Take only x, y, z
                    point_clouds.append(point_cloud)
            except Exception as e:
                print(f"Skipping frame due to parsing error: {e}")

    return point_clouds

def visualize_frames(point_clouds):
    vis = o3d.visualization.Visualizer()
    vis.create_window()
    
    if not point_clouds:
        print("No valid frames to visualize.")
        return

    pcd = point_clouds[0]  # Start with the first frame
    vis.add_geometry(pcd)

    for i, pcd in enumerate(point_clouds):
        print(f"Displaying frame {i+1}/{len(point_clouds)}")
        vis.clear_geometries()
        vis.add_geometry(pcd)
        vis.poll_events()
        vis.update_renderer()

    vis.destroy_window()

# Example usage
file_path = r"C:\Users\HP\Downloads\M.pcd"  # Change this to your actual PCD file path
point_clouds = load_multiple_pcd_frames(file_path)
visualize_frames(point_clouds)
```

```python
import open3d as o3d
import glob
import time
import numpy as np

# Get all PCD files in sorted order (change the path as needed)
pcd_files = sorted(glob.glob(r"C:\Users\HP\Downloads\pcd\M1_pcd_data\*.pcd"))

# Create a visualizer
vis = o3d.visualization.Visualizer()
vis.create_window(window_name="Point Cloud Viewer", width=1280, height=720)  # Set window size

# Function to preprocess PCD files
def preprocess_pcd(file):
    pcd = o3d.io.read_point_cloud(file)

    # 1️⃣ Downsampling - Reduces points while keeping structure
    pcd = pcd.voxel_down_sample(voxel_size=0.01)  # Adjust voxel size as needed

    # 2️⃣ Noise Removal - Removes random noise
    #pcd, _ = pcd.remove_statistical_outlier(nb_neighbors=30, std_ratio=2.0)

    # 3️⃣ Plane Segmentation (RANSAC) - Removes dominant planes like ground
    # plane_model, inliers = pcd.segment_plane(distance_threshold=0.02,
    #                                          ransac_n=3,
    #                                          num_iterations=1000)
    # inlier_cloud = pcd.select_by_index(inliers, invert=True)  # Remove ground
    # pcd = inlier_cloud  # Keep only non-ground points

    # 4️⃣ Clustering (DBSCAN) - Removes small noise clusters
    labels = np.array(pcd.cluster_dbscan(eps=0.05, min_points=10, print_progress=False))
    largest_cluster = max(labels) if len(labels) > 0 else -1
    filtered_pcd = pcd.select_by_index(np.where(labels == largest_cluster)[0])

    return filtered_pcd

# Initialize point cloud with first frame
pcd = preprocess_pcd(pcd_files[0])
vis.add_geometry(pcd)

# Loop through all frames and update visualization
for file in pcd_files:
    pcd_new = preprocess_pcd(file)  # Apply preprocessing to each frame
    pcd.points = pcd_new.points
    pcd.colors = pcd_new.colors

    vis.update_geometry(pcd)
    vis.poll_events()
    vis.update_renderer()
    time.sleep(0.5)  # Adjust frame delay for smoother animation

vis.destroy_window()
```

```python
import open3d as o3d
import glob
import time
import numpy as np

# Get all PCD files in sorted order (change the path as needed)
pcd_files = sorted(glob.glob(r"C:\Users\HP\Downloads\pcd\M1_pcd_data\*.pcd"))

# Create a visualizer
vis = o3d.visualization.Visualizer()
vis.create_window()

# Create a point cloud object for updating
pcd = o3d.io.read_point_cloud(pcd_files[0])
vis.add_geometry(pcd)

# Function to detect objects and compute bounding boxes
def detect_objects_and_compute_bboxes(pcd):
    # Perform clustering to detect objects
    labels = np.array(pcd.cluster_dbscan(eps=0.05, min_points=10, print_progress=False))

    # Get the number of objects (clusters)
    max_label = labels.max()
    print(f"Detected {max_label + 1} objects in the point cloud.")

    # List to store bounding boxes
    bboxes = []

    # Iterate through each cluster and compute bounding boxes
    for label in range(max_label + 1):
        # Get points belonging to the current cluster
        cluster_indices = np.where(labels == label)[0]
        cluster_points = pcd.select_by_index(cluster_indices)

        # Compute the oriented bounding box (OBB)
        bbox = cluster_points.get_oriented_bounding_box()

        # Color the bounding box (optional)
        bbox.color = [1, 0, 0]  # Red color

        # Add the bounding box to the list
        bboxes.append(bbox)

    return bboxes

# Loop through all frames and update visualization
for file in pcd_files:
    # Read the new point cloud
    new_pcd = o3d.io.read_point_cloud(file)

    # Update the point cloud data
    pcd.points = new_pcd.points
    pcd.colors = new_pcd.colors  # Update colors (if available)

    # Detect objects and compute bounding boxes
    bboxes = detect_objects_and_compute_bboxes(pcd)

    # Clear previous bounding boxes (if any)
    for geometry in vis.get_geometries(): -- # get geometries is out of current version of Open3d
        if isinstance(geometry, o3d.geometry.OrientedBoundingBox):
            vis.remove_geometry(geometry, reset_bounding_box=False)

    # Add new bounding boxes to the visualizer
    for bbox in bboxes:
        vis.add_geometry(bbox, reset_bounding_box=False)

    # Update visualization
    vis.update_geometry(pcd)
    vis.poll_events()
    vis.update_renderer()
    time.sleep(1)  # Adjust frame delay for smoother animation

# Keep the window open until it is closed by the user
vis.run()

# Destroy the window
vis.destroy_window()
```
